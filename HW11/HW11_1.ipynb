```python
import math
import numpy as np
from sympy import *
import matplotlib.pyplot as plt

x = symbols ('x')
y = -12*x**4 - 18*x**3+5*x**2 + 10*x - 30
y
```

```python
# Определить корни
print(solve(y)[0])
print(solve(y)[1])
print(solve(y)[2])
print(solve(y)[3])
```

```python
#Построить график
p= plot(y, (x, -3, 2))
```

```python
#Найти интервалы, на которых функция возрастает
y = -12*x**4 - 18*x**3+5*x**2 + 10*x - 30
dx = diff(y)
print(solve(dx > 0))
```

```python
# Найти интервалы, на которых функция убывает
print(solve(dx < 0))
```

```python
# Вычислить вершины
apex = solve(dx)
print (apex[0])
print (apex[1])
print (apex[2])
```
